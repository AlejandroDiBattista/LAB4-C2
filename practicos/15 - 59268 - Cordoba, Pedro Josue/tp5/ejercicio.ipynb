## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python


def verificarCumplimientoDeRangos(pCod, pNom, pPrec, pTip, pCant):
    pasoLaPrueba = True

    if(pCod != 4) and (len(pNom) < 1 or len(pNom) > 100)and (pPrec < 10 or pPrec > 10000) and (len(pTip) < 0 or len(pTip) > 100) and (pCant < 0 or pCant > 100):
        pasoLaPrueba = False

    return pasoLaPrueba


class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):   
            
        if verificarCumplimientoDeRangos(codigo, nombre, precio, tipo, cantidad):
            self._codigo = codigo
            self._nombre = nombre
            self._precio = precio
            self._tipo = tipo
            self._cantidad = cantidad
       

    @property
    def codigo(self):
        return self._codigo
    
    @codigo.setter
    def codigo(self, pCod):
        if pCod == 4:
            self._codigo = pCod


    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, pNom):
        if len(pNom) > 0 and len(pNom) <= 100:
            self._nombre = pNom

    @property
    def precio(self):
        return self._precio
    
    
    @precio.setter
    def precio(self, pPrec):
        if pPrec >= 10 and pPrec <= 10000:
            self._precio = pPrec

    @property
    def cantidad(self):
        return self._cantidad
    
    @cantidad.setter
    def cantidad(self, pCant):
        if pCant >= 0 and pCant <= 100:
            self._cantidad = pCant


    @property
    def tipo(self):
        return self._tipo
    
    @tipo.setter
    def tipo(self, pTipo):
        self._tipo = pTipo

    
    def valorTotal(self):
        return self._precio * self._cantidad
    
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv

class Catalogo:

    list_datos = []
    datos = []
    archivo = "catalogo-prueba.csv"
    archivoProductos = "productos.csv"
    lector = ""

    instanciasOferta = [];
    descuento = 0

    productos = []
    ofertaDescuento = []
    oferta2X1 = []
    
    

    def __init__(self) : pass

    #Reconstruir el objeto mediante la obtencion de datos del archivo csv
    def leer(pArchivo):
        catalogo = Catalogo();
        catalogo.productos = []

        file = open(pArchivo, mode='r', newline= '')
        lector = csv.reader(file)

        for l in lector:
            producto = Producto(int(l[0]), l[1], float(l[2]), l[3], int(l[4]))
            catalogo.agregar(producto)
            
        return catalogo


    def guardar(self,archivo):
        file = open(archivo, mode='w', newline= '')
        escrbir = csv.writer(file)

        for p in self.productos:
            escrbir.writerow([p.codigo, p.nombre, p.precio, p.tipo, p.cantidad])

        file.close();
    

    def agregar(self, *producto):
        for p in producto:
            self.productos.append(p)
            

    def buscar(self,codigo) : 
        resultado = None
        for p in self.productos:
            if p.codigo == codigo:
                resultado = p
        return resultado
                

    def informe(self): 
        resultado = f"INFORME CATALOGO \n    Cantidad de productos: {str(self.cantidadProductos)}\n   Cantidad de unidades: {str(self.cantidadUnidades)}\n  Precio Promedio: {str(self.precioPromedio)}\n    Valor total: {self.valorTotal}\n Tipos de productos: \n {self.imprimirTiposDeProductos} \n Ofertas: \n Oferta 2x1 "
        return resultado

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def valorTotal(self):
        resultado = 0
        for p in self.productos:
            resultado += p.precio * p.cantidad

        return resultado;
    

    @property
    def cantidadUnidades(self):
        resultado = 0
        for p in self.productos:
            resultado += p.cantidad;

        return resultado
    
    @property
    def precioPromedio(self):
        return self.valorTotal / self.cantidadProductos
    
    @property
    def imprimirTiposDeProductos(self):
        resultado = ""
        for p in self.productos:
            resultado += f"-Tipo: {p.tipo} : Unidades {str(p.cantidad)} : Precio promedio $ \n"

        return resultado
    

    def calcularDescuento(self,producto, unidades):
         resultado = 0
         #esta variable guarda una comparacion entre descuentos
         masFavorable = 0
         for of in self.instanciasOferta:
             resultado = of.calcularDescuento(producto,unidades)
             if resultado >= masFavorable:
                 masFavorable = resultado

         return masFavorable
    
        
    def registrarOferta(self,oferta):
        self.instanciasOferta.append(oferta);
    

    def vender(self,producto, cantidad):
        for p in self.productos:
            if p == producto and p.cantidad >= cantidad:
                p.cantidad -= cantidad;
                
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
def esCorrectoElNombre(pNombre):
    resultado = False

    if(pNombre != "" and len(pNombre) > 0 and len(pNombre) <= 100):
        resultado = True
    return resultado

def esCorrectoElCuit(pCuit):
    resultado = True
    listCuit = list(pCuit);
    contador = 0

    if len(listCuit) == 13 and listCuit[2] == '-' and listCuit[11] == '-':
        while contador < len(listCuit):
            if contador != 2 and contador != 11:
                if not listCuit[contador].isnumeric():
                    resultado = False;
            contador += 1
    else:
        resultado = False

    return resultado
            
        

class Cliente: 
    def __init__(self, nombre, cuit): 
        if esCorrectoElNombre(nombre) and esCorrectoElCuit(cuit):
            self._nombre = nombre
            self._cuit = cuit
        else:
            print("No es posible crear la instancia, datos incorrectos!")


    @property
    def nombre(self):
        return self._nombre;

    @nombre.setter
    def nombre(self, pNombre):
        if esCorrectoElNombre(pNombre):
            self._nombre = pNombre;

    @property
    def cuit(self):
        return self._cuit;

    @cuit.setter
    def cuit(self, pCuit):
        if esCorrectoElCuit(pCuit):
            self._cuit = pCuit;

```

```python
class Factura:
    _numero = 100  # Inicializa el número de la factura en 100
    
    def __init__(self, catalogo, cliente):
        self.catalogo = catalogo
        self.cliente = cliente
        self.items = []  # Lista de productos comprados
        self.numero = Factura.nuevoNumero()  # Asigna un nuevo número de factura
        self.descuentos = 0  # Inicializa descuentos
        self.subtotal = 0  # Inicializa subtotal

    @classmethod
    def nuevoNumero(cls):
        cls._numero += 1
        return cls._numero

    @classmethod
    def ultimaFactura(cls, numero):
        cls._numero = numero

    def agregar(self, producto, cantidad):
        # Busca el producto en el catálogo
        prod_catalogo = self.catalogo.buscar(producto.codigo)

        if prod_catalogo and prod_catalogo.cantidad >= cantidad:
            # Verifica si el producto ya está en los items
            for item in self.items:
                if item['producto'].codigo == producto.codigo:
                    # Si el producto ya existe, solo actualiza la cantidad y el subtotal
                    item['cantidad'] += cantidad
                    item['subtotal'] += prod_catalogo.precio * cantidad
                    break
            else:
                # Si no está en los items, se agrega como nuevo
                subtotal_producto = prod_catalogo.precio * cantidad
                self.items.append({'producto': producto, 'cantidad': cantidad, 'subtotal': subtotal_producto})

            # Actualiza el stock del catálogo
            self.catalogo.vender(prod_catalogo, cantidad)

            # Aquí está la corrección
            if 'item' in locals():
                # Actualiza el subtotal solo si el producto ya estaba en items
                self.subtotal += prod_catalogo.precio * cantidad  # Solo sumar el nuevo subtotal
            else:
                # Si es un nuevo producto, simplemente suma el subtotal del nuevo producto
                self.subtotal += subtotal_producto

            # Calcula el descuento
            self.aplicarDescuentos(producto, cantidad)
        else:
            print("Producto no disponible o cantidad insuficiente.")

    def aplicarDescuentos(self, producto, cantidad):
        # Reiniciar descuentos antes de calcular
        self.descuentos = 0
        for oferta in self.catalogo.instanciasOferta:
            if isinstance(oferta, OfertaDescuento):
                if producto.codigo in oferta.codigos:
                    descuento = oferta.calcularDescuento(producto, cantidad)
                    self.descuentos += descuento
    
            elif isinstance(oferta, Oferta2x1):
                if producto.codigo in oferta.codigos or producto.tipo in oferta.tipos:
                    descuento = oferta.calcularDescuento(producto, cantidad)
                    self.descuentos += descuento

    @property
    def total(self):
        # Calcula el total final después de aplicar descuentos
        return self.subtotal - self.descuentos

    @property
    def cantidadProductos(self):
        return len(self.items)

    @property
    def cantidadUnidades(self):
        return sum(item['cantidad'] for item in self.items)

    def imprimir(self):
        # Genera la impresión de la factura
        informe = f"Factura Número: {self.numero}\n"
        informe += f"Cliente: {self.cliente.nombre}\n"
        informe += "Productos:\n"
        for item in self.items:
            informe += f"- {item['producto'].nombre}: {item['cantidad']} x ${item['producto'].precio} = ${item['subtotal']:.2f}\n"

        if self.descuentos > 0:
            informe += f"Descuentos: -${self.descuentos:.2f}\n"
        
        informe += f"TOTAL: ${self.total:.2f}\n"
        return informe
```
